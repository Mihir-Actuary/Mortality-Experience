## Data **Analysis**

Populations from Census

```r

Census_Data_C = Census_Data_C[-(122:124), -10]
colnames(Census_Data_C) = c("Age", "1996 Male Population", "1996 Female Population",
                                   "2001 Male Population", "2001 Female Population",
                                   "2011 Male Population", "2011 Female Population",
                                   "Total Male Population", "Total Female Population")

Age_Gender_Population_1996 = data.frame(Census_Data_C$Age, Census_Data_C$`1996 Male Population`, Census_Data_C$`1996 Female Population`)
Age_Gender_Population_1996$'Total Population' = Age_Gender_Population_1996$Census_Data_C..1996.Male.Population. + Age_Gender_Population_1996$Census_Data_C..1996.Female.Population.

Age_Gender_Population_2001 = data.frame(Census_Data_C$Age, Census_Data_C$`2001 Male Population`, Census_Data_C$`2001 Female Population`)
Age_Gender_Population_2001$'Total Population' = Age_Gender_Population_2001$Census_Data_C..2001.Male.Population. + Age_Gender_Population_2001$Census_Data_C..2001.Female.Population.

Age_Gender_Population_2011 = data.frame(data.frame(Census_Data_C$Age, Census_Data_C$`2011 Male Population`, Census_Data_C$`2011 Female Population`))
Age_Gender_Population_2011$'Total Population' = Age_Gender_Population_2011$Census_Data_C..2011.Male.Population. + Age_Gender_Population_2011$Census_Data_C..2011.Female.Population.

Age_Gender_Population_2022 = Age_Gender_Population_2022[-(122:134), -5]
colnames(Age_Gender_Population_2022) = c("Age", "Male Population", "Female Population", "Total Population")

###   Population Growth Model(Using Splines)

Census_Years = c(1996, 2001, 2011, 2022)
Ages = 1:121

Males_P_Matrix = cbind(Census_Data_C$Age,
                     Age_Gender_Population_1996$Census_Data_C..1996.Male.Population.,
                     Age_Gender_Population_2001$Census_Data_C..2001.Male.Population.,
                     Age_Gender_Population_2011$Census_Data_C..2011.Male.Population.,
                     Age_Gender_Population_2022$`Male Population`)

Females_P_Matrix = cbind(Census_Data_C$Age,
                       Age_Gender_Population_1996$Census_Data_C..1996.Female.Population.,
                       Age_Gender_Population_2001$Census_Data_C..2001.Female.Population.,
                       Age_Gender_Population_2011$Census_Data_C..2011.Female.Population.,
                       Age_Gender_Population_2022$`Female Population`)

Totals_P_Matrix = cbind(Census_Data_C$Age,
                      Age_Gender_Population_1996$`Total Population`,
                      Age_Gender_Population_2001$`Total Population`,
                      Age_Gender_Population_2011$`Total Population`,
                      Age_Gender_Population_2022$`Total Population`)

Interpolated_P_Years = 1996:2022

Interpolated_P_Males = matrix(NA, nrow = length(Ages), ncol = length(Interpolated_P_Years))
Interpolated_P_Females = matrix(NA, nrow = length(Ages), ncol = length(Interpolated_P_Years))
Interpolated_P_Totals = matrix(NA, nrow = length(Ages), ncol = length(Interpolated_P_Years))

for(i in 1 : length(Ages)) {
  Pop_Vals_M = as.numeric(Males_P_Matrix[i, 2:5])
  Spline_Model_M = lm(Pop_Vals_M ~ ns(Census_Years, df = 3))
  Interpolated_P_Males[i, ] = predict(Spline_Model_M, newdata = data.frame(Census_Years = Interpolated_P_Years))
  colnames(Interpolated_P_Males) = Interpolated_P_Years
  rownames(Interpolated_P_Males) = Ages
}

for(i in 1 : length(Ages)) {
  Pop_Vals_F = as.numeric(Females_P_Matrix[i, 2:5])
  Spline_Model_F = lm(Pop_Vals_F ~ ns(Census_Years, df = 3))
  Interpolated_P_Females[i, ] = predict(Spline_Model_F, newdata = data.frame(Census_Years = Interpolated_P_Years))
  colnames(Interpolated_P_Females) = Interpolated_P_Years
  rownames(Interpolated_P_Females) = Ages
}

for(i in 1 : length(Ages)) {
  Pop_Vals_T = as.numeric(Totals_P_Matrix[i, 2:5])
  Spline_Model_T = lm(Pop_Vals_T ~ ns(Census_Years, df = 3))
  Interpolated_P_Totals[i, ] = predict(Spline_Model_T, newdata = data.frame(Census_Years = Interpolated_P_Years))
  colnames(Interpolated_P_Totals) = Interpolated_P_Years
  rownames(Interpolated_P_Totals) = Ages
}
```

```r

Age_Gender_1997_2015 = Age_Gender_1997_2015[-(122:136), -5]
Age_Gender_1997_2016 = Age_Gender_1997_2016[-(122:136), -5]
Age_Gender_1997_2017 = Age_Gender_1997_2017[-(122:136), -5]
Age_Gender_1997_2018 = Age_Gender_1997_2018[-(122:136), -5]
Age_Gender_1997_2019 = Age_Gender_1997_2019[-(122:136), -5]
Age_Gender_1997_2020 = Age_Gender_1997_2020[-(122:136), -5]

Age_Gender_2016 = data.frame(Age_Gender_1997_2016[ , 2:4] - Age_Gender_1997_2015[ , 2:4])
Age_Gender_2016 = cbind(Age = 0:120, Age_Gender_2016)
colnames(Age_Gender_2016) = c("Age", "Male Deaths", "Female Deaths", "Total Deaths")

Age_Gender_2017 = data.frame(Age_Gender_1997_2017[ , 2:4] - Age_Gender_1997_2016[ , 2:4])
Age_Gender_2017 = cbind(Age = 0:120, Age_Gender_2017)
colnames(Age_Gender_2017) = c("Age", "Male Deaths", "Female Deaths", "Total Deaths")

Age_Gender_2018 = data.frame(Age_Gender_1997_2018[ , 2:4] - Age_Gender_1997_2017[ , 2:4])
Age_Gender_2018 = cbind(Age = 0:120, Age_Gender_2018)
colnames(Age_Gender_2018) = c("Age", "Male Deaths", "Female Deaths", "Total Deaths")

Age_Gender_2019 = data.frame(Age_Gender_1997_2019[ , 2:4] - Age_Gender_1997_2018[ , 2:4])
Age_Gender_2019 = cbind(Age = 0:120, Age_Gender_2019)
colnames(Age_Gender_2019) = c("Age", "Male Deaths", "Female Deaths", "Total Deaths")

Age_Gender_2020 = data.frame(Age_Gender_1997_2020[ , 2:4] - Age_Gender_1997_2019[ , 2:4])
Age_Gender_2020 = cbind(Age = 0:120, Age_Gender_2020)
colnames(Age_Gender_2020) = c("Age", "Male Deaths", "Female Deaths", "Total Deaths")

Age_Gender_2007 = data.frame(Age = 0:120,
                              `Male Deaths` = Age_Gender_2006$`Male Deaths` + (1/2) * (Age_Gender_2008$`Male Deaths` - Age_Gender_2006$`Male Deaths`),
                              `Female Deaths` = Age_Gender_2006$`Female Deaths` + (1/2) * (Age_Gender_2008$`Female Deaths` - Age_Gender_2006$`Female Deaths`),
                              `Total Deaths` = Age_Gender_2006$`Total Deaths` + (1/2) * (Age_Gender_2008$`Total Deaths` - Age_Gender_2006$`Total Deaths`))
#Age_Gender_2007 = round(Age_Gender_2007[ , 2:4])
Age_Gender_2007 = cbind(Age = 0:120, Age_Gender_2007)
colnames(Age_Gender_2007) = c("Age", "Male Deaths", "Female Deaths", "Total Deaths")

Interpolated_D_Years = 2006:2022
Death_Years = c(2006:2021)

Males_D_Matrix = cbind(Census_Data_C$Age,
                       Age_Gender_2006$`Male Deaths`,
                       Age_Gender_2007$`Male Deaths`,
                       Age_Gender_2008$`Male Deaths`,
                       Age_Gender_2009$`Male Deaths`,
                       Age_Gender_2010$`Male Deaths`,
                       Age_Gender_2011$`Male Deaths`,
                       Age_Gender_2012$`Male Deaths`,
                       Age_Gender_2013$`Male Deaths`,
                       Age_Gender_2014$`Male Deaths`,
                       Age_Gender_2015$`Male Deaths`,
                       Age_Gender_2016$`Male Deaths`,
                       Age_Gender_2017$`Male Deaths`,
                       Age_Gender_2018$`Male Deaths`,
                       Age_Gender_2019$`Male Deaths`,
                       Age_Gender_2020$`Male Deaths`,
                       Age_Gender_2021$`Male Deaths`)

Females_D_Matrix = cbind(Census_Data_C$Age,
                        Age_Gender_2006$`Female Deaths`,
                        Age_Gender_2007$`Female Deaths`,
                        Age_Gender_2008$`Female Deaths`,
                        Age_Gender_2009$`Female Deaths`,
                        Age_Gender_2010$`Female Deaths`,
                        Age_Gender_2011$`Female Deaths`,
                        Age_Gender_2012$`Female Deaths`,
                        Age_Gender_2013$`Female Deaths`,
                        Age_Gender_2014$`Female Deaths`,
                        Age_Gender_2015$`Female Deaths`,
                        Age_Gender_2016$`Female Deaths`,
                        Age_Gender_2017$`Female Deaths`,
                        Age_Gender_2018$`Female Deaths`,
                        Age_Gender_2019$`Female Deaths`,
                        Age_Gender_2020$`Female Deaths`,
                        Age_Gender_2021$`Female Deaths`)

Totals_D_Matrix = cbind(Census_Data_C$Age,
                        Age_Gender_2006$`Total Deaths`,
                        Age_Gender_2007$`Total Deaths`,
                        Age_Gender_2008$`Total Deaths`,
                        Age_Gender_2009$`Total Deaths`,
                        Age_Gender_2010$`Total Deaths`,
                        Age_Gender_2011$`Total Deaths`,
                        Age_Gender_2012$`Total Deaths`,
                        Age_Gender_2013$`Total Deaths`,
                        Age_Gender_2014$`Total Deaths`,
                        Age_Gender_2015$`Total Deaths`,
                        Age_Gender_2016$`Total Deaths`,
                        Age_Gender_2017$`Total Deaths`,
                        Age_Gender_2018$`Total Deaths`,
                        Age_Gender_2019$`Total Deaths`,
                        Age_Gender_2020$`Total Deaths`,
                        Age_Gender_2021$`Total Deaths`)

Interpolated_D_Males = matrix(NA, nrow = length(Ages), ncol = length(Interpolated_D_Years))
Interpolated_D_Females = matrix(NA, nrow = length(Ages), ncol = length(Interpolated_D_Years))
Interpolated_D_Totals = matrix(NA, nrow = length(Ages), ncol = length(Interpolated_D_Years))

for(i in 1 : length(Ages)) {
  Pop_Vals_M_D = as.numeric(Males_D_Matrix[i, 2:17])
  Spline_Model_M_D = lm(Pop_Vals_M_D ~ ns(Death_Years, df = 6))
  Interpolated_D_Males[i, ] = predict(Spline_Model_M_D, newdata = data.frame(Death_Years = Interpolated_D_Years))
  colnames(Interpolated_D_Males) = Interpolated_D_Years
  rownames(Interpolated_D_Males) = Ages
}

for(i in 1 : length(Ages)) {
  Pop_Vals_F_D = as.numeric(Females_D_Matrix[i, 2:17])
  Spline_Model_F_D = lm(Pop_Vals_F_D ~ ns(Death_Years, df = 6))
  Interpolated_D_Females[i, ] = predict(Spline_Model_F_D, newdata = data.frame(Death_Years = Interpolated_D_Years))
  colnames(Interpolated_D_Females) = Interpolated_D_Years
  rownames(Interpolated_D_Females) = Ages
}

for(i in 1 : length(Ages)) {
  Pop_Vals_T_D = as.numeric(Totals_D_Matrix[i, 2:17])
  Spline_Model_T_D = lm(Pop_Vals_T_D ~ ns(Death_Years, df = 6))
  Interpolated_D_Totals[i, ] = predict(Spline_Model_T_D, newdata = data.frame(Death_Years = Interpolated_D_Years))
  colnames(Interpolated_D_Totals) = Interpolated_D_Years
  rownames(Interpolated_D_Totals) = Ages
}
```
